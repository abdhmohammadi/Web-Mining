```python
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd
from urllib.request import urlopen
```

Setting the site address and reading the page

```python
url = 'https://ur.isc.ac/Home/RankIranUniv'

page = urlopen(url)

html_bytes = page.read()

html_string = html_bytes.decode("utf-8")
```

Generating a dataframe from data

```python
data_ = []

soup = BeautifulSoup(html_string, 'html.parser')

tr_list = soup.findAll('tr')

i = 1

for item in tr_list: 
    
    str_item = str(item)

    soup = BeautifulSoup(str_item,'html.parser')

    td_list = soup.findAll('td')
    
    if len(td_list)>1:
    
      # Removeing html charachters
      td_value = str(re.sub('<.*?>','',str(td_list[2])))
      # removeing extra line break charachters
      td_value = td_value.replace('\\n','')
      td_value = td_value.replace('\\r','')
      td_value = td_value.strip()

      row = [str(i), td_value]
      
      data_.append(row)

      i+=1

df = pd.DataFrame(columns=['Rank','University'], data=data_)

df.head(int(len(df)*0.2))
```

Generating a HTML and README.MD output

```python
# Saving the dataframe into the html file
df.to_html(buf= 'Iran-ISC-Ranking.html', header=True,index=False, justify='center',encoding='utf-8')

# Read the html file to generate README.MD
htmlfile = open('Iran-ISC-Ranking.html',mode='r',encoding='utf-8')

# Get HTML string
html_string = htmlfile.read()

htmlfile.close()

# Generate Readme file for github
readme_text = "<div align=center dir=rtl><h2>Scrapping Iranian universities ranking</h2><br/>"\
              "https://ur.isc.ac/Home/RankIranUniv</div><br/>\n" \
              "<b>By: Abdullah Mohammadi<br/>abdhmohammady@gmail.com</b><br/><b>https://github.com/AbdhMohammadi</b><br/>"\
              "<b>June, 10, 2023</b><br/><br/>"\

readme_text = readme_text + '\n'+ html_string + "<h5>Please give me a star if this is usefull code.</h5>"
# Create or open README.MD
readmefile = open('README.MD','w',encoding='utf-8')

readmefile.write(readme_text)

readmefile.flush()

readmefile.close()

print('HTML output and Readme file was created.')
```
